```python
import chromadb
client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about Birmingham",
        "This document is about LA"
    ],
    ids = ['id1','id2']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
document = collection.get(ids=["id1"])
document
```

```python
results = collection.query(
    query_texts=['Query is about California beach'],
    n_results = 2
)
results
```
